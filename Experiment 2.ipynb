### Experiment 2

To plot various types of Activation functions used in artificial Neural Networks

#### Importing required Libraries

```python
import numpy  as np
import matplotlib.pyplot as plt
```

#### Binary Step Function

Binary Step function return 0 if the value is less then 0 , return 1 If the value is more then 0

```python
def binaryStep(x):
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10,10)
plt.plot(x,binaryStep(x))
plt.axis('tight')
plt.title('Activation Function : binary step')
plt.show()
```

#### Linear Activation Function

```python
def linear(x):
    return x
```

```python
x = np.linspace(-10,10)
plt.plot(x,linear(x))
plt.axis('tight')
plt.title('Activation function: Linear')
plt.show()
```

#### Sigmoid Function

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10,10)
plt.plot(x,sigmoid(x))
plt.axis('tight')
plt.title('Activation Function  :Sigmoid')
plt.show()
```

#### Tanh Function

```python
def tanh(x):
    #return the value (1-exp(-2x))/(1+exp(-2x)) which lies in the range -1 and 1 
    return np.tanh(x)
```

```python
x = np.linspace(-10,10)
plt.plot(x,tanh(x))
plt.axis('tight')
plt.title('Activation Function: TanH')
plt.show()
```

#### Relu Function

```python
def RELU(x):
    x1=[]
    for i in x:
        if i<0:
            x1.append(0)
        else:
            x1.append(i)
    return x1
```

```python
x = np.linspace(-10,10)
plt.plot(x,RELU(x))
plt.axis('tight')
plt.title('Activation function: ReLU Function')
plt.show()
```

#### Softmax Activation Function

```python
def softmax(x):
    return np.exp(x)/np.sum(np.exp(x), axis=0)
```

```python
x = np.linspace(-10,10)
plt.plot(x, softmax(x))
plt.axis('tight')
plt.title('Activation function: Softmax Function')
plt.show()
```

#### Leaky Relu Function

```python
def leaky_relu(x):
    data = [max(0.05*value, value) for value in x]
    return np.array(data, dtype=float)
```

```python
x = np.linspace(-10,10)
plt.plot(x,leaky_relu(x))
plt.axis('tight')
plt.title('Activation function: LeakyReLu')
plt.show()
```

#### Bipolar Sigmoid Function

```python
def bipolarsigmoid(x):
    return (-1) + (2/(1+np.exp(-x)))
```

```python
x = np.linspace(-10,10)
plt.plot(x,bipolarsigmoid(x))
plt.axis('tight')
plt.title('Activation function: BipolarSigmoid')
plt.show()
```
